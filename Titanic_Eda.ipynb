## Selecting and Filtering

```python
# Select the "Name" column.
df[['Name']]
```

```python
# Select "Name", "Sex", and "Age" columns.

df[['Name','Sex','Age']]
```

```python
# Filter passengers older than 60.

df[df['Age']>60]
```

```python
# Filter passengers who paid fare greater than $100.
df[df['Fare'] > 100]
```

```python
# Filter female passengers.

df[df['Sex']=='female']
```

```python
# Select passengers who are female and older than 50.
df[(df['Sex']=='female' ) & (df['Age']>50)]
```

```python
# Select passengers with Cabin info not missing.
df[df['Cabin'].notna()]
```

```python
# Select passengers who embarked from ‘S’.
df[df['Embarked']=='S']
```

```python
# Use .iloc to get the first row.
df.iloc[0]
```

```python
# Use .loc to get the row with index 5.

df.loc[5]
```
