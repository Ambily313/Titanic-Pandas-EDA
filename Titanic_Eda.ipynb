### Data Cleaning

```python
# Replace missing 'Age' values with median.

df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
# Fill missing 'Embarked' with the most common value.

df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)

```

```python
# Drop the 'Cabin' column.

df.drop(columns='Cabin',inplace=True)
```

```python
# Drop rows where 'Embarked' is missing.

df.dropna(subset=['Embarked'],inplace=True)
```

```python
# Convert 'Sex' column to lowercase.

df['Sex']=df['Sex'].str.lower()
df.sample(5)
```

```python
# Strip white spaces in column names.

df.columns=df.columns.str.strip()
df.columns
```

```python
# Rename the 'Fare' column to 'TicketPrice'.

df.rename(columns={'Fare':'TicketPrice'},inplace=True)
df.sample(3)
```

###  Transformations

```python
# Create a new column 'FamilySize' = SibSp + Parch + 1.

df['FamilySize']=df['SibSp']+df['Parch']+1
df.sample(3)
```

```python
# Create a new column 'IsMinor' for Age < 18.

df['IsMinor']=df['Age'] < 18
df.sample(3)
```

```python
#  Map 'Sex' to numbers (male: 0, female: 1).

df['Sex']=df['Sex'].map({'male':0,'female':1})

df.sample(3)
```

```python
# Normalize the 'Fare' column.

df['FareNorm']=(df['TicketPrice']-df['TicketPrice'].min())/(df['TicketPrice'].max()-df['TicketPrice'].min())
df.sample(5)
```

```python
# Bin the 'Age' column into categories: Child, Teen, Adult, Senior.

df['Agegruop']=pd.cut(df['Age'],bins=[0,12,18,60,100],labels=['Child','Teen', 'Adult','Senior'])
df.sample(10)
```

### Boolean and Membership

```python
# Find passengers with names containing 'Smith'.

df[df['Name'].str.contains('Smith',case=False,na=False)]
```

```python
# Select rows where 'Ticket' starts with '3'.
df[df['Ticket'].str.startswith('3',na=False)]
```

```python
# Select passengers whose 'Age' is between 20 and 40.
# df[(df['Age']>=20) & (df['Age']<=40)]

df[df['Age'].between(20, 40)]
```

```python
# Select passengers who embarked from 'C' or 'Q'.

# df[(df['Embarked']=='C') | (df['Embarked']=='Q') ]

df[df['Embarked'].isin(['C', 'Q'])]
```

```python
# Check if any passenger is older than 80.
# df[df['Age']>80]

(df['Age'] > 80).any()
```

### Visualization

#### 1. Survival Count Plot (by Gender)

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival Count by Gender')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.xlabel('Gender')
plt.ylabel('Passenger Count')
plt.show()
```

##### Insight: Much higher survival rate among females compared to males

#### 2.Age Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Age'].dropna(),bins=30,kde=True)
plt.title("Age Distribution of Passangers")
plt.xlabel('Age')
plt.ylabel('Frequency')
```

##### Insight: Most passengers were between 20â€“40 years old.

### 3.Survival by Passenger Class

```python
sns.countplot(data=df,x='Pclass',hue='Survived')
plt.title("Survival Count by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Count")
```

#### Insight: 1st class passengers had a higher survival rate.

### 4.Boxplot of Fare by Class

```python
sns.boxplot(data=df, x='Pclass', y='TicketPrice')
plt.title('Fare Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
```

##### Insight:Higher classes paid significantly more on average.

### 5.Embarked vs Survival Heatmap

```python
Embarked_survival=pd.crosstab(df['Embarked'],df['Survived'])
sns.heatmap(Embarked_survival,annot=True,cmap='Blues')
plt.title("Survival by Embarked Port")
plt.xlabel('Survived')
plt.ylabel('Embarked Station')
```

##### Insight: Passengers from 'C' had a higher proportion of survivors.

### 6.Age vs Fare (with Survival Hue)

```python
sns.scatterplot(data=df, x='Age', y='TicketPrice', hue='Survived')
plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
```

#### Insight: Some very young and elderly passengers paid high fares â€” mostly in 1st class.
